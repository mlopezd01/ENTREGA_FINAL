#### PROYECTO: SERIES Y PELÍCULAS DE LA PLATAFORMA NETFLIX

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### ANÁLISIS EDA

```python
df = pd.read_csv('https://raw.githubusercontent.com/GabiMiranda05/data-science-portfolio/refs/heads/main/Ejercicios-python/netflix_titles.csv', encoding = 'latin-1')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Eliminar columnas desde Unnamed: 12 a Unnamed: 25
df_copia = df.copy()
df_copia = df_copia.drop(df_copia.columns[12:26], axis=1)
df_copia.head()
```

Se eliminan las columnas del dataset ya que no contienen información en su totalidad y los nombres de las columnas tampoco indican que debiera tener contenido.

```python
# Se calcula la moda en la columna rating
moda_rating = df_copia['rating'].mode()[0]
print(moda_rating)
```

```python
# Completar rating con la moda de la columna
df_copia['rating'] = df_copia['rating'].fillna(moda_rating)
```

```python
# Convertimos a string y limpiamos espacios
df_copia['date_added'] = df_copia['date_added'].astype(str).str.strip()

# Convertimos a datetime, forzando errores como NaT
df_copia['date_added'] = pd.to_datetime(df_copia['date_added'], errors='coerce')

# Extraemos año y mes
df_copia['year_added'] = df_copia['date_added'].dt.year.astype('Int64')
df_copia['month_added'] = df_copia['date_added'].dt.month.astype('Int64')
```

```python
df_copia['duration'].unique()
```

```python
# Limpieza de columna duración
df_copia['duration_int'] = df_copia['duration'].str.extract('(\d+)').astype(float)
df_copia['duration_type'] = df_copia['duration'].str.extract('([a-zA-Z]+)')
```

```python
df_copia.head()
```

```python
df_copia = df_copia.dropna(subset=['duration'])
```

```python
# Calcular el porcentaje de valores nulos por columna en el DataFrame
porcentaje_nulos_df = df_copia.isnull().mean() * 100
porcentaje_nulos_df
```

```python
# Completar los valores nulos
df_copia['director'] = df_copia['director'].fillna('Desconocido')
df_copia['cast'] = df_copia['cast'].fillna('Desconocido')
df_copia['country'] = df_copia['country'].fillna('Desconocido')
```

```python
# Se elimnan los registros sin fecha ya que se consideran poco representativos
df_copia.dropna(subset=['date_added'], inplace=True)
```

```python
# Calcular nulos
df_copia.isnull().sum()
```

```python
# Gráfico barras de Cantidad de peliculas adheridas a Netflix VS Cantidad de Series
# Conteo series
sns.countplot(data=df_copia, x='type')
plt.title('Cantidad de películas y series en Netflix')
plt.xlabel('Tipo')
plt.ylabel('Cantidad')
plt.show()
```

```python
df_copia['release_year_group'] = np.where(df_copia['release_year'] < 2012, 'Antes_de_2012', df_copia['release_year'])
```

```python
df_copia['release_year_group'].unique()
```

```python
# Gráfico de barras de Cantidad de peliculas realizadas en función del año
Conteo_years = df_copia['release_year_group'].value_counts()
plt.pie(Conteo_years.values, labels=Conteo_years.index, autopct='%1.1f%%')
plt.title('Cantidad de películas lanzadas por año')
plt.show()
```

```python
# DISTRIBUACIÓN DE RATINGS (CLASIFICACIÓN POR EDAD)
plt.figure(figsize=(10, 5))
df_copia['rating'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 ratings más frecuentes")
plt.ylabel("Cantidad")
plt.xlabel("Rating")
plt.show()
```

```python
# EVOLUCIÓN DEL CONTENIDO SEGÚN EL TIEMPO
df_copia['release_year'].value_counts().sort_index().plot(kind="line", figsize=(12, 5))
plt.title("Cantidad de títulos lanzados por año")
plt.ylabel("Número de títulos")
plt.xlabel("Año")
plt.grid()
plt.show()
```

```python
# TENDENCIA MENSUAL DE TÍTULOS AÑADIDOS POR MES
plt.figure(figsize=(12, 5))
sns.countplot(data=df_copia, x='month_added')
plt.title('Títulos añadidos por mes')
plt.xlabel('Mes')
plt.ylabel('Cantidad de títulos')
plt.tight_layout()
plt.show()
```

```python
# DURACIÓN SEGÚN TIPO
plt.figure(figsize=(8, 5))
sns.boxplot(x="type", y="duration_int", data=df_copia)
plt.title("Distribución de duración por tipo")
plt.ylabel("Duración (minutos o temporadas)")
plt.show()
```

```python
# DISTRIBUCIÓN DE DURACIÓN (PELÍCULAS)
movies = df_copia[df_copia['type'] == 'Movie']

plt.figure(figsize=(10, 5))
sns.histplot(movies['duration_int'], bins=30, kde=True, color='skyblue')
plt.title('Distribución de duración de películas (en minutos)')
plt.xlabel('Duración (min)')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

```python
# GÉNEROS MÁS COMUNES
from collections import Counter

# Contar los géneros (listed_in puede tener múltiples por fila)
genres = df_copia['listed_in'].dropna().str.split(', ')
flat_genres = [genre for sublist in genres for genre in sublist]
genre_counts = Counter(flat_genres)

# Top 10
top_genres = dict(genre_counts.most_common(10))

# Gráfico de barras
plt.figure(figsize=(10, 5))
sns.barplot(x=list(top_genres.values()), y=list(top_genres.keys()), hue=list(top_genres.keys()), palette="viridis", legend=False)
plt.title("Top 10 géneros más comunes")
plt.xlabel("Cantidad de títulos")
plt.ylabel("Género")
plt.show()
```

```python
# RELACIÓN ENTRE RATING Y TIPO DE CONTENIDO
pd.crosstab(df_copia['rating'], df_copia['type']).plot(kind="bar", stacked=True, figsize=(10,6), colormap="coolwarm")
plt.title("Distribución de ratings por tipo de contenido")
plt.xlabel("Rating")
plt.ylabel("Cantidad")
plt.show()
```

```python
top_countries = df_copia['country'].value_counts().head(10)
```

```python
# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 10 países con más títulos en Netflix', fontsize=14)
plt.xlabel('Cantidad de títulos')
plt.ylabel('País')
plt.tight_layout()
plt.show()
```

El gráfico muestra que Estados Unidos lidera ampliamente en cantidad de títulos en Netflix, seguido por India y Reino Unido. Esto refleja el peso de sus industrias audiovisuales y la estrategia de Netflix de producir contenido local en mercados clave. También destacan países como Corea del Sur, Japón y España, lo que muestra la apuesta por la diversidad cultural en su catálogo.

```python
# Agrupar por país y tipo de contenido
country_type_counts = df_copia.groupby(['country', 'type']).size().reset_index(name='count')

# Seleccionar los 6 países con más títulos totales
top_countries_list = df_copia['country'].value_counts().head(6).index
filtered_data = country_type_counts[country_type_counts['country'].isin(top_countries_list)]

# Gráfico
plt.figure(figsize=(10, 6))
sns.barplot(data=filtered_data, x='count', y='country', hue='type', palette='pastel')
plt.title('Distribución de Movies y TV Shows en países con más títulos')
plt.xlabel('Cantidad de títulos')
plt.ylabel('País')
plt.legend(title='Tipo de contenido')
plt.tight_layout()
plt.show()
```
